```python
import numpy as np
import matplotlib.pyplot as plt

# set width of bar
barWidth = 0.25
fig = plt.subplots(figsize =(16, 8))

# set height of bar
no_quant = [8.037, 7.008, 5.756, 4.423, 2.612, 0.917]
fp16_quant = [4.03, 3.614, 3.089, 2.429, 1.66, 0.84]
dynamic_quant = [2.031, 1.354, 1.162, 0.942, 0.672, 0.323]

# Set posno_quantion of bar on X axis
br1 = np.arange(len(no_quant))
br2 = [x + barWidth for x in br1]
br3 = [x + barWidth for x in br2]

# Make the plot
plt.bar(br1, no_quant, color= 'w', width = barWidth,
		edgecolor ='black', label ='No Quantization', hatch= '/')
plt.bar(br2, fp16_quant, width = barWidth, color='w',
		edgecolor ='black', label ='Floating Point 16 Quantization', hatch='*')
plt.bar(br3, dynamic_quant, color ='w', width = barWidth,
		edgecolor ='black', label ='8 bit Integer Quantization', hatch='o')

# Adding Xticks
plt.xlabel('Sparsity %', fontweight ='bold', fontsize = 25)
plt.ylabel('Model Size (MB)', fontweight ='bold', fontsize = 25)
plt.xticks([r + barWidth for r in range(len(no_quant))],
		['0', '10', '30', '50', '70', '90'], fontsize=25)
plt.yticks(fontsize=25)

plt.legend(fontsize=20)
plt.show()
```
